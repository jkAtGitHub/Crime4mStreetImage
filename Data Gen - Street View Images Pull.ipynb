```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import json
import requests
from PIL import Image
from io import BytesIO 
```

```python
df = pd.read_csv('train.csv')
df.head()
```

```python
import os
```

```python
def merge_images(image1, image2, image3, image4):
    result = Image.new('RGB', (400, 400))
    result.paste(im=image1, box=(0, 0))
    result.paste(im=image2, box=(200, 0))
    result.paste(im=image3, box=(0, 200))
    result.paste(im=image4, box=(200, 200))
    return result
```

```python
key = "AIzaSyByVbYYSIflZfib3TXbucZkRrITnjML070"
for i, row in df.iterrows():

    lon=row['x']
    lat=row['y']

    data=requests.get(f"https://maps.googleapis.com/maps/api/streetview?size=200x200&location={lat},{lon}&fov=90&heading=0&pitch=10%20&key={key}")
    file_like = BytesIO(data.content)
    img1 = Image.open(file_like)
    #Heading = 90
    data=requests.get(f"https://maps.googleapis.com/maps/api/streetview?size=200x200&location={lat},{lon}&fov=90&heading=90&pitch=10%20&key={key}")

    file_like = BytesIO(data.content)
    img2 = Image.open(file_like)
    #Heading = 180
    data=requests.get(f"https://maps.googleapis.com/maps/api/streetview?size=200x200&location={lat},{lon}&fov=90&heading=180&pitch=10%20&key={key}")

    file_like = BytesIO(data.content)
    img3 = Image.open(file_like)
    #Heading = 270
    data=requests.get(f"https://maps.googleapis.com/maps/api/streetview?size=200x200&location={lat},{lon}&fov=90&heading=270&pitch=10%20&key={key}")
    file_like = BytesIO(data.content)
    img4 = Image.open(file_like)
    result_img=merge_images(img1,img2,img3,img4)
    
    result_img.save(f"./Images/{row['image_name']}.jpg")
                    
                    
print("Complete")
```
